```python
Q-1:
    In NumPy, there is a difference between a variable declared as a Python list and a NumPy array.

A Python list is a built-in data structure in Python that can contain elements of different types. 
It is a flexible container that can be resized and modified after creation. However, Python lists 
are not optimized for numerical computations and do not provide many mathematical functions.

On the other hand, a NumPy array is a central data structure in the NumPy library. 
It is a homogeneous, multidimensional container for elements of the same type,
typically numbers. NumPy arrays are designed for efficient numerical operations 
and provide a wide range of mathematical functions and operations optimized for
performance. NumPy arrays have a fixed size upon creation and cannot be resized directly; 
instead, a new array needs to be created with the desired size.


import numpy as np

# Python list
py_list = [1, 2, 3, 4, 5]

# NumPy array
np_array = np.array([1, 2, 3, 4, 5])

print(type(py_list))   # <class 'list'>
print(type(np_array))  # <class 'numpy.ndarray'>

```

```python
#Q-2:
import numpy as np

# Variable list
var_list = [1, 2.5, 'hello', True]

# Variable array
var_array = np.array([1, 2.5, 'hello', True])

# Print data types of elements in the list
print("Data types of elements in the variable list:")
for element in var_list:
    print(type(element))

# Print data types of elements in the array
print("\nData types of elements in the variable array:")
for element in var_array:
    print(type(element))
```

```python
#Q-3:
import numpy as np

# Original list
list_ = [1, 2.5, 'hello', True]

# Convert list to NumPy array with data type 'int'
arr_list = np.array(object=list_, dtype=int)

# Print data types of elements in the original list
print("Data types of elements in the original list:")
for element in list_:
    print(type(element))

# Print data types of elements in the NumPy array
print("\nData types of elements in the NumPy array:")
for element in arr_list:
    print(type(element))
# As you can see, the original list contains elements of
different types (int, float, str, bool). However, when 
the list is converted to a NumPy array with a data type 
of int, all elements are converted to numpy.int64, which 
the specified data type.
```
